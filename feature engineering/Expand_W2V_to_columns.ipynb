```python
import pandas as pd
from sqlalchemy import create_engine

engine_out = create_engine(r'postgresql+psycopg2://postgres:stats170@104.197.51.5')
```

```python
df = pd.read_sql_query("select * from yelp_data.yelp_vertorize", con = engine_out)
df
```

```python
temp = df['vertorize_review'].to_list()
temp = [i[1:-1].split(',') for i in temp]
temp = pd.DataFrame(temp)
temp.columns = ['class_'+str(i) for i in temp.columns]
```

```python
df = pd.concat((df,temp), axis=1)
df.to_sql('yelp_vectorize', con = engine_out, if_exists = 'replace', index = True,schema = 'yelp_data')
```
