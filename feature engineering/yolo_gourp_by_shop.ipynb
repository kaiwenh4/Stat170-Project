```python
import pandas as pd
from sqlalchemy import create_engine

engine_out = create_engine(r'postgresql+psycopg2://postgres:stats170@104.197.51.5')
```

```python
df = pd.read_sql_query("SELECT * FROM yelp_data.yelp_yolo", con = engine_out)
df
```

```python
# photo id and business_id are incorrect
# photo id is business_id, and business_id is photo_id
# group by and sum all features
df = df.groupby("photo_id").sum()
# correct the index name
df = df.rename_axis("business_id")
# delete photo_id column
df = df.loc[:, df.columns!='photo_index']
```

```python
# to_Sql
df.to_sql('yelp_yolo_by_shop', con = engine_out, if_exists = 'replace', index = True,schema = 'yelp_data')
```
